```python
#1. Get the numerical variables from our dataset.
customer_df_num = customer_df.select_dtypes(include = np.number)
customer_df_num.head()
```

```python
#2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
for col in customer_df_num.columns:
    sns.displot(customer_df_num[col],kde=True)
    plt.show()
```

# After analysis, it is apparent that a number of columns, including income, monthly premium auto, and total claim amount, show traits of either partial or full exponential distributions. We'll use a log transform to determine whether these distributions can be normalized.

# Customer lifetime value and total claim amount appear to be the most closely aligned with an exponential distribution of all the columns.
# Despite the fact that the latter is an output variable. 
# We will use the following adjustments to see if these columns can be normalized.

```python
def log_trans(x):
    transformed_x = np.log10(x)
    return transformed_x if np.isfinite(transformed_x) else np.NAN
```

```python
def quantile_trans(df, col_name):
    iqr = np.percentile(df[col_name], 75) - np.percentile(df[col_name], 25)
    upper_limit = np.percentile(df[col_name], 75) + 1.5 * iqr
    lower_limit = np.percentile(df[col_name], 25) - 1.5 * iqr
    return df[(df[col_name] > lower_limit) & (df[col_name] < upper_limit)]
```

```python
def boxcox_trans(df, col_name):
    df[col_name] = np.where(df[col_name] <= 0, 0, df[col_name])
    return stats.boxcox(df[col_name])[0]
```

```python
for col in ['customer_lifetime_value', 'total_claim_amount']:
    # Log Transform
    print(f"Log Transform: {col}")
    customer_df_num[col+'_log'] = np.log10(customer_df_num[col])
    customer_df_num[col+'_log'] = customer_df_num[col+'_log'].fillna(np.mean(customer_df_num[col]))
    sns.histplot(customer_df_num[col+'_log'], kde=True)
    plt.show()
```
